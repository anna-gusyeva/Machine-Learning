```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = "/content/drive/My Drive/bigml_59c28831336c6604c800002a.csv"
bnk = pd.read_csv(path, sep=',')
bnk.head()
```

```python
print(bnk['churn'].value_counts(normalize=True))
normalized_counts = bnk['churn'].value_counts(normalize=True)

plt.figure(figsize=(6, 4))
normalized_counts.plot(kind='bar')
plt.title('Churn Value Counts')
plt.xlabel('Churn')
plt.ylabel('Percentage')
plt.xticks(rotation=0)
plt.show()
```

```python
bnk.drop(["phone number","area code"], axis = 1, inplace=True)
bnk.head(10)
```

```python
bnk['international plan'] = bnk['international plan'].map({'no': 0, 'yes': 1})
bnk['voice mail plan'] = bnk['voice mail plan'].map({'no': 0, 'yes': 1})
bnk['churn'] = bnk['churn'].map({False: 0, True: 1})
bnk.head(10)
```

```python
bnk = pd.get_dummies(bnk, columns=['state'], prefix=['state'], drop_first=True)
bnk.head(10)
```

```python
bnk.info()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

std_col = ['account length', 'number vmail messages', 'total day minutes', 'total day calls', 'total day charge', 'total eve minutes',
                   'total eve calls', 'total eve charge', 'total night minutes', 'total night calls', 'total night charge', 'total intl minutes',
                   'total intl calls', 'total intl charge', 'customer service calls']

bnk[std_col] = scaler.fit_transform(bnk[std_col])
bnk.head()
```

```python
from sklearn.model_selection import train_test_split
X = bnk.drop('churn', axis=1)
y = bnk['churn']
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size = 0.25, random_state=3)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)

X_train_st = scaler.transform(X_train)
X_valid_st = scaler.transform(X_valid)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import f1_score, accuracy_score, recall_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

#Ð¡Ñ‚Ð²Ð¾Ñ€ÐµÐ½Ð½Ñ Ñ‚Ð° Ð½Ð°Ð²Ñ‡Ð°Ð½Ð½Ñ Ð±Ð°Ð³Ð°Ñ‚Ð¾ÑˆÐ°Ñ€Ð¾Ð²Ð¾Ð³Ð¾ Ð¿ÐµÑ€Ñ†ÐµÐ¿Ñ‚Ñ€Ð¾Ð½Ñƒ
NeuralNet = MLPClassifier(random_state=5, max_iter=1000)
NeuralNet.fit(X_train_st, y_train)


y_pred = NeuralNet.predict(X_valid_st)

f1 = f1_score(y_valid, y_pred)
accuracy = accuracy_score(y_valid, y_pred)
recall = recall_score(y_valid, y_pred)

df_scores = pd.DataFrame({'Neural network': [f1, accuracy, recall]}, index=['f1_score', 'accuracy_score', 'recall_score'])
print("ÐœÐµÑ‚Ñ€Ð¸ÐºÐ¸ ÑÐºÐ¾ÑÑ‚Ñ– Ð½Ð° Ð²Ð°Ð»Ñ–Ð´Ð°Ñ†Ñ–Ð¹Ð½Ð¸Ñ… Ð´Ð°Ð½Ð¸Ñ…:")
print(df_scores)

conf_matrix = confusion_matrix(y_valid, y_pred)
print("\nÐœÐ°Ñ‚Ñ€Ð¸Ñ†Ñ Ð¿Ð»ÑƒÑ‚Ð°Ð½Ð¸Ð½Ð¸:")
print(conf_matrix)

sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False, annot_kws={'size': 14})
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()

class_report = classification_report(y_valid, y_pred)
print("\nÐ—Ð²Ñ–Ñ‚ Ð¿Ñ€Ð¾ ÐºÐ»Ð°ÑÐ¸Ñ„Ñ–ÐºÐ°Ñ†Ñ–ÑŽ:")
print(class_report)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.neural_network import MLPClassifier
from scipy.stats import randint

#Ð’Ð¸Ð·Ð½Ð°Ñ‡ÐµÐ½Ð½Ñ Ð¿Ñ€Ð¾ÑÑ‚Ð¾Ñ€Ñƒ Ð³Ñ–Ð¿ÐµÑ€Ð¿Ð°Ñ€Ð°Ð¼ÐµÑ‚Ñ€Ñ–Ð²
param_dist = {
    'hidden_layer_sizes': [(randint.rvs(10, 200),) for _ in range(10)],
    'activation': ['logistic', 'tanh', 'relu'],
    'alpha': [10 ** i for i in range(-6, 1)],
    'learning_rate_init': [10 ** i for i in range(-4, 0)]
}

neural_net = MLPClassifier(random_state=5, max_iter=1000)

random_search = RandomizedSearchCV(neural_net, param_distributions=param_dist, n_iter=20, cv=5, scoring='f1', random_state=42, n_jobs=-1)
random_search.fit(X_train_st, y_train)

print("ÐšÑ€Ð°Ñ‰Ñ– Ð¿Ð°Ñ€Ð°Ð¼ÐµÑ‚Ñ€Ð¸:", random_search.best_params_)
print("ÐšÑ€Ð°Ñ‰Ð° Ð¾Ñ†Ñ–Ð½ÐºÐ° (F1-score):", random_search.best_score_)
```

```python
from sklearn.model_selection import validation_curve

param_range = param_dist['hidden_layer_sizes']
train_scores, test_scores = validation_curve(neural_net, X_train_st, y_train, param_name='hidden_layer_sizes', param_range=param_range, cv=5, scoring='f1', n_jobs=-1)

plt.figure(figsize=(10, 6))
plt.plot(param_range, np.mean(train_scores, axis=1), label='Training Score')
plt.plot(param_range, np.mean(test_scores, axis=1), label='Validation Score')
plt.xlabel('Hidden Layer Sizes')
plt.ylabel('F1 Score')
plt.title('Validation Curve for Hidden Layer Sizes')
plt.legend()
plt.show()
```

```python
y_pred = random_search.best_estimator_.predict(X_valid_st)

f1 = f1_score(y_valid, y_pred)
recall = recall_score(y_valid, y_pred)
accuracy = accuracy_score(y_valid, y_pred)

print('f1 score:', f1)
print('recall:', recall)
print('accuracy:', accuracy)
```

```python
conf_matrix = confusion_matrix(y_valid, y_pred)
print("\nÐœÐ°Ñ‚Ñ€Ð¸Ñ†Ñ Ð¿Ð»ÑƒÑ‚Ð°Ð½Ð¸Ð½Ð¸:")
print(conf_matrix)

sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False, annot_kws={'size': 14})
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.linear_model import LogisticRegression

baggingLogReg = BaggingClassifier(LogisticRegression(), n_estimators=20, max_samples=0.7, max_features=0.7, random_state=5)
baggingLogReg.fit(X_train_st, y_train)

y_pred = baggingLogReg.predict(X_valid_st)

print('f1 score:', f1_score(y_pred, y_valid, average='macro'))
print('recall:', recall_score(y_pred, y_valid, average='macro'))
print('accuracy:', accuracy_score(y_pred, y_valid))
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV, KFold

base_model = LogisticRegression(solver='liblinear')
baggingLogReg = BaggingClassifier(base_model, random_state=42)

kf = KFold(n_splits=5, shuffle=True, random_state=42)

param_grid = {'n_estimators': [10, 20, 30, 40, 60, 80, 100]}

baggingLogReg_grid = GridSearchCV(baggingLogReg, param_grid, cv=kf, scoring='f1_macro', n_jobs=-1)

baggingLogReg_grid.fit(X_train_st, y_train)

print("ÐšÑ€Ð°Ñ‰Ñ– Ð¿Ð°Ñ€Ð°Ð¼ÐµÑ‚Ñ€Ð¸:", baggingLogReg_grid.best_params_)
print("Ð ÐµÐ·ÑƒÐ»ÑŒÑ‚Ð°Ñ‚ (f1_macro):", baggingLogReg_grid.best_score_)
```

```python
y_pred = baggingLogReg_grid.best_estimator_.predict(X_valid_st)

print('f1 score:', f1_score(y_pred, y_valid, average='macro'))
print('recall:', recall_score(y_pred, y_valid, average='macro'))
print('accuracy:', accuracy_score(y_pred, y_valid))
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer, f1_score, accuracy_score, recall_score
import matplotlib.pyplot as plt

base_model = DecisionTreeClassifier(max_depth=3)
abTree = AdaBoostClassifier(base_model, random_state=5)

param_grid = {'n_estimators': [50, 100, 150, 200, 250]}

abTree_grid = GridSearchCV(abTree, param_grid, cv=5, scoring=make_scorer(f1_score, average='macro'),n_jobs=-1)

abTree_grid.fit(X_train_st, y_train)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(param_grid['n_estimators'], abTree_grid.cv_results_['mean_test_score'], marker='o')
plt.title('Ð’Ð°Ð»Ñ–Ð´Ð°Ñ†Ñ–Ð¹Ð½Ð° ÐºÑ€Ð¸Ð²Ð° Ð´Ð»Ñ AdaBoost')
plt.xlabel('ÐšÑ–Ð»ÑŒÐºÑ–ÑÑ‚ÑŒ Ð´ÐµÑ€ÐµÐ² Ñƒ Ð±ÑƒÑÑ‚Ð¸Ð½Ð³Ñƒ (n_estimators)')
plt.ylabel('Ð¡ÐµÑ€ÐµÐ´Ð½Ñ” Ð·Ð½Ð°Ñ‡ÐµÐ½Ð½Ñ F1_macro Ð½Ð° Ð²Ð°Ð»Ñ–Ð´Ð°Ñ†Ñ–Ñ—')
plt.show()
```

```python
y_pred = abTree_grid.predict(X_valid_st)

print('f1 score:', f1_score(y_pred, y_valid, average='macro'))
print('recall:', recall_score(y_pred, y_valid, average='macro'))
print('accuracy:', accuracy_score(y_pred, y_valid))
```
